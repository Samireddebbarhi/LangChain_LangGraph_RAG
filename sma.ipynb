### Dynamic Model

```python
from langchain.agents.middleware import wrap_model_call, ModelRequest, ModelResponse
from langchain.messages import AIMessage, HumanMessage, SystemMessage
from IPython.display import display, Markdown   
```

```python
advanced_llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash",temperature=0)
basic_llm = ChatOllama(model="gemma4:e4b",temperature=0)
```

```python
"""
@parametre request :Contient toutes les informations de l'appel.
     request.model: modèle actuellement utilisé
     request.runtime: contexte d'exécution de l'appel
     request.override: fonction pour remplacer le modèle ou d'autres paramètres de l'appel
     request.messages: messages d'entrée de l'appel
@parametre handler :fonction qui exécute l'appel du modèle avec les paramètres actuels ou modifiés
@return ModelResponse : réponse du modèle après l'exécution de handler
"""
@wrap_model_call
def daynamic_model_selection(request: ModelRequest, handler)-> ModelResponse:
   evn = request.runtime.context.get("env", "test")
   if evn == "prod":
        request.model = advanced_llm
        print("advanced_llm selected ")
   else:
        request.model = basic_llm
        print("basic_llm selected ")
   return handler (request.override(model=request.model))
```

```python
agent_with_middleware = create_agent(
    model=basic_llm,  # model par defaut, sera remplacé par le middleware
    tools=[],
    middleware=[daynamic_model_selection],
    debug=True
)
```

```python
response=agent_with_middleware.invoke(input={"messages":[
    HumanMessage(content="C est quoi un agent AI ?")]},
    context={"env": "test"}
)
```

```python
print(display(Markdown(response['messages'][-1].content)))
```

## Tools

```python
from langchain.tools import tool
```

```python
@tool
def get_meteo(city:str):
    """ 
    Get wheather information for a given city.
    """
    print("Weather tool called with city:", city)
    return {
        "city": city,
        "temperature": 25,
        "humidity": 80,
        "pressure": 103
    }

@tool
def get_employee_info(employee_name :str):
    """ 
    Get info about the given employee (Salary, Seniority).
    """
    print("get_employee_info invoked")
    return {
        "name": employee_name,
        "salary": 25000,
        "seniority": 5,
    }
```

```python
agent4 = create_agent(
    model=advanced_llm,
    system_prompt="respond to the user query and use the tools if needed",
    tools=[get_meteo, get_employee_info],
    checkpointer=memory
)
```

```python
config={"configurable": {"thread_id":1}}
resp=agent4.invoke(
    input={"messages":[HumanMessage(content="What is the weather in Paris ?")]},
    config=config
    )
```

```python
print(resp['messages'][-1].content)
```

```python
resp=agent4.invoke(
    input={"messages":[HumanMessage(content="quel est lesalaire de Hassan?")]},
    config=config
    )
```

```python
print(resp['messages'][-1].content)
```

### Tools (DuckDuckGo)

```python
from ddgs import DDGS
@tool
def web_search(query: str, num_results:int=5) -> str:
    """
    Search the web usin DuckDuckGo
    Args:
    query : Search query string
    num_results : Number of results to return (Default=5)
    Returns:
    Formatted search results with titles, descptions and Urls
    """
    print(f'Web search tool called with query: {query} ')
    try:
        print(f'Search tool is called for {query}')
        ddgs_search = DDGS()
        results=ddgs_search.text(query=query, max_results=num_results, backend="google")
        if not results:
            return f"No results found for {query} "
        formatted_results = [f"Search for {query} : \n"]
        for i, result in enumerate(results,1):
            title = result.get("title","No Title")
            body = result.get("body","No description available")
            href = result.get("href","")
            formatted_results.append( f"{i}. **{title}**. {body}. {href}")
            return formatted_results
    except Exception as e:
        print(str(e))
```

```python
agent5= create_agent(
    model=basic_llm,
    system_prompt="respond to the user query and use the tools if needed",
    tools=[get_meteo, get_employee_info, web_search],
    checkpointer=memory
)
```

```python
config={"configurable": {"thread_id":1}}
resp=agent5.invoke(
    input={"messages":[HumanMessage(content="donne moi les liens des sites pour aprrendre python")]},
    config=config
    )
```

```python
print(display(Markdown(resp['messages'][-1].content)))
```

### Tools (TAVILY search)

```python
load_dotenv(override=True)
```

```python
from langchain_tavily import TavilySearch
```

```python
tavily= TavilySearch(max_results=10,search_depth="advanced")
```

```python
@tool
def search_web(query: str,max_results:int=10) :
    """
    Search for general current web results like time in a city, news, etc... using Tavily API
    """
    print(f'Search tool is called for {query}')
    result=tavily.invoke({"query":query})
    return result
```

```python
agent5= create_agent(
    model=advanced_llm,
    system_prompt="respond to the user query and use the tools if needed",
    tools=[get_meteo, get_employee_info, search_web],
    checkpointer=memory
)
```

```python
config={"configurable": {"thread_id":1}}
resp1=agent5.invoke(
    input={"messages":[HumanMessage(content="cherche moi dans les webs actualité quel heure fait-il actuellementà casbalanca Morocco ?")]},
    config=config
    )
```
